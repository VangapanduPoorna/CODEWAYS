```python
import random

# Define the animal-related questions and answers
animal_questions = [
    {
        "question": "What is the largest mammal in the world?",
        "answer": "blue whale"
    },
    {
        "question": "What is the fastest land animal?",
        "answer": "cheetah"
    },
    {
        "question": "What is the only continent without any native reptiles or snakes?",
        "answer": "antarctica"
    },
    {
        "question": "Which bird is known for its long neck and legs, and is also the national bird of Tanzania?",
        "answer": "ostrich"
    },
    {
        "question": "What is the only mammal capable of sustained flight?",
        "answer": "bat"
    },
    {
        "question": "What is the largest living primate?",
        "answer": "gorilla"
    },
    {
        "question": "Which insect is the only one capable of turning its head 360 degrees?",
        "answer": "praying mantis"
    },
    {
        "question": "What is the only mammal capable of true sustained flight?",
        "answer": "bat"
    },
    {
        "question": "What is the largest living land animal?",
        "answer": "african elephant"
    },
    {
        "question": "What animal is known as the 'ship of the desert'?",
        "answer": "camel"
    },
    {
        "question": "Which bird lays the largest eggs relative to its body size?",
        "answer": "kiwi"
    },
    {
        "question": "What is the only marsupial native to North America?",
        "answer": "virginia opossum"
    },
    {
        "question": "What animal is the national symbol of Australia?",
        "answer": "kangaroo"
    },
    {
        "question": "What is the only bird that can fly backward?",
        "answer": "hummingbird"
    },
    {
        "question": "Which sea creature is known for its ability to change color and camouflage itself?",
        "answer": "octopus"
    },
    {
        "question": "Which animal is known for its distinctive black and white stripes and is native to Africa?",
        "answer": "zebra"
    },
    {
        "question": "What is the world's largest reptile?",
        "answer": "saltwater crocodile"
    },
    {
        "question": "Which big cat is the fastest swimmer?",
        "answer": "tiger"
    },
    {
        "question": "What animal has the longest tongue relative to its body size?",
        "answer": "chameleon"
    },
    {
        "question": "Which animal is known as the 'King of the Jungle'?",
        "answer": "lion"
    }
]

def display_question(question_data):
    print(question_data["question"])

def evaluate_answer(user_answer, correct_answer):
    if user_answer.lower() == correct_answer.lower():
        print("Correct!")
        return 1
    else:
        print("Incorrect. The correct answer is:", correct_answer)
        return 0

def main():
    print("Welcome to the Animal Quiz!")
    print("You will be asked animal-related questions. Enter your answer.")
    total_questions = len(animal_questions)
    score = 0

    # Shuffle the questions to make it more challenging
    random.shuffle(animal_questions)

    for question_data in animal_questions:
        display_question(question_data)
        user_answer = input("Enter your answer: ")
        score += evaluate_answer(user_answer, question_data["answer"])

    print("Quiz completed!")
    print("Your final score:", score, "out of", total_questions)

    play_again = input("Do you want to play again? (yes/no): ").lower()
    if play_again == "yes":
        main()
    else:
        print("Thank you for playing!")

if __name__ == "__main__":
    main()
```
